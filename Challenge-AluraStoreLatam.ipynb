### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
IngresoTienda1 = tienda['Precio'].sum()
IngresoTienda2 = tienda2['Precio'].sum()
IngresoTienda3 = tienda3['Precio'].sum()
IngresoTienda4 = tienda4['Precio'].sum()
print(f"El ingreso de la tienda 1 es: {IngresoTienda1}")
print(f"El ingreso de la tienda 2 es: {IngresoTienda2}")
print(f"El ingreso de la tienda 3 es: {IngresoTienda3}")
print(f"El ingreso de la tienda 4 es: {IngresoTienda4}")
```

```python
##Creando un DataFrame para crear los gráfico
df_ingresos = pd.DataFrame({'Tiendas' : ['tienda','tienda2','tienda3','tienda4'], 'Ingresos' : [IngresoTienda1,IngresoTienda2,IngresoTienda3,IngresoTienda4]})
df_ingresos
```

```python
import matplotlib.pyplot as plt
```

```python
df_ingresos['Porcentajes'] = df_ingresos['Ingresos'] / df_ingresos['Ingresos'].sum() * 100
df_ingresos
```

```python
df_ingresos.plot(kind='bar',x='Tiendas',y='Porcentajes',title='Porcentaje de Ingresos por Tienda',legend=False)
```

```python
df_ingresos.plot(kind='pie',y='Porcentajes',labels=df_ingresos['Tiendas'],autopct='%1.1f%%',startangle=90,legend=False,title='Distribución Porcentual de Ingresos por Tienda')
```

# 2. Ventas por categoría

####En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
producT1=tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
producT2=tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
producT3=tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
producT4=tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

# 4. Productos más y menos vendidos
